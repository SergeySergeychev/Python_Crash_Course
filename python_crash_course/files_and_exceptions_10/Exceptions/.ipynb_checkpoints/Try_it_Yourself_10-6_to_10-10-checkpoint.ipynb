```python
# Code for EX 10-6 Addition:

try:
    number_one = input("Enter a first number: ")
    int_number_one = int(number_one)
    
    number_two = input("Enter a second number: ")
    int_number_two = int(number_two)
    
except ValueError:
    msg = "Invalid number, please write only numericals."
    print(msg)
        
else:
    sum_of_numbers = int_number_one + int_number_two
    print("The sum of two numbers is equal: " + str(sum_of_numbers))
```

```python
# EX 10-7 Addition Calculator:

print("\nIf you want to quit enter 'q'.\n")

while True:
    number_one = input("Enter a first number: ")
    if number_one == 'q':
        break
    
    number_two = input("Enter a second number: ")
    if number_two == 'q':
        break
    try:        
        answer = int(number_one) + int(number_two)
    except ValueError:
        msg = "\nInvalid number, please enter numerical."
        print(msg)
    else:
        print("\nYour sum of numbers is equal to " + str(answer))
        
    
    
```

```python
# EX 10-10. Common Words:
# Condition:
#    Visit project Gutenberg and analyze text.

# Task:
#    Visit https://gutenberg.org and find text.
#    Count how many time the word the appears in text files string.


# Test: 
#    Test program and use text.lower() to count all the words

# Solution:    
    # Download text from https://gutenberg.org.
    # Save at name_of_a_book.txt
    # Read name_of_a_book.txt
    # Use function to count the the word in  name_of_a_book.txt.
    # Print the result on desktop.
    
# Code from ehmatthes.github.io

def count_common_words(filename, word):
    """Count how many times word appears in the text."""
    # Note: This is a really simplea approximation, and the number returned
    # will be higher than the actual count.
    
    try:
        with open(filename, encoding='utf8') as f:
            contents = f.read()
    except FileNotFoundError:
        pass
    
    else:
        word_count = contents.lower().count(word)
        
        msg = f"'{word}' appears in {filename} about {word_count} times."
        print(msg)
        
filename = 'alice.txt'
count_common_words(filename,'the')
```
